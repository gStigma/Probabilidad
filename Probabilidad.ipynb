## 2. Probabilidad de eventos


La probabilidad de un evento  $\Omega$ denotada como $P(\Omega)$, mide la fracción de veces que esperaríamos que ocurra el evento si el experimento se repitiera muchas veces bajo las mismas condiciones. En términos simples, la probabilidad es un número entre 0 y 1, donde:


- $P(\Omega)$ =0 indica que el evento nunca ocurre.
- $P(\Omega)$ =1 indica que el evento siempre ocurre.
- Si 0<$P(\Omega)$<1, entonces el evento ocurre con cierta probabilidad intermedia.

## 3.	Probabilidad Condicional.

La **probabilidad condicional** es la probabilidad de que ocurra un evento \(A\), dado que se sabe que ocurrió otro evento \(B\). Se denota como \(P(A \mid B)\) y se calcula como:

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}
$$

Esto es útil para analizar situaciones en las que el conocimiento de un evento influye en la probabilidad de otro, donde se puede leer como la probabilidad de que un resultado sea favorable para el evento \(A\) dado que el evento \(B\) fue observado.


La definición de **probabilidad condicional** también se puede reorganizar para obtener la regla del producto:

$$
P(A \cap B) = P(B) \cdot P(A \mid B)
$$

Esto dice que la probabilidad de que ocurran ambos eventos \(A\) y \(B\) es igual a la probabilidad de \(B\) multiplicada por la probabilidad de que ocurra \(A\), dado que \(B\) ha ocurrido.


### Ejemplo 3.1

```python
import numpy as np

# Supongamos que tenemos dos eventos:
# Evento A: "Sacar un número par al lanzar un dado"
# Evento B: "Sacar un número mayor que 3 al lanzar un dado"

# Definimos los eventos
total_lanzamientos = 6  # Total de resultados posibles al lanzar un dado

# Probabilidad de A (números pares: 2, 4, 6)
numeros_pares = [2, 4, 6]
P_A = len(numeros_pares) / total_lanzamientos

# Probabilidad de B (números mayores que 3: 4, 5, 6)
numeros_mayores_que_3 = [4, 5, 6]
P_B = len(numeros_mayores_que_3) / total_lanzamientos

# Probabilidad de A y B (números pares que son mayores que 3: 4, 6)
numeros_pares_y_mayores_que_3 = [4, 6]
P_A_interseccion_B = len(numeros_pares_y_mayores_que_3) / total_lanzamientos

# Calcular P(A | B) usando la fórmula de probabilidad condicional
P_A_dado_B = P_A_interseccion_B / P_B

# Imprimir resultados
print(f"P(A): {P_A:.2f}")
print(f"P(B): {P_B:.2f}")
print(f"P(A ∩ B): {P_A_interseccion_B:.2f}")
print(f"P(A | B): {P_A_dado_B:.2f}")

# Verificación de la regla del producto
P_A_interseccion_B_verificacion = P_B * P_A_dado_B
print(f"P(A ∩ B) usando la regla del producto: {P_A_interseccion_B_verificacion:.2f}")
```

## 4.	Regla de la cadena.

La **regla de la cadena** es un método que nos permite descomponer una probabilidad conjunta de múltiples eventos en términos de probabilidades condicionales.

Si \(A\) y \(B\) son independientes, entonces:

$$
P(A \cap B) = P(A) P(B)
$$

Esto implica que

$$
P(B \mid A) = P(B)
$$

Por lo tanto, también

$$
P(A \mid B) = P(A)
$$

Cuando los eventos son dependientes, la probabilidad conjunta y las condicionales cambian en función de la relación entre los eventos.


### Ejemplo 4.1


```python
# Supongamos que tenemos dos eventos:
# Evento A: "Llover hoy"
# Evento B: "Llevar paraguas"

# Definimos las probabilidades:
P_A = 0.3  # Probabilidad de que llueva hoy
P_B_dado_A = 0.9  # Probabilidad de llevar paraguas dado que llueve
P_B_dado_no_A = 0.2  # Probabilidad de llevar paraguas dado que no llueve

# Calcular P(B) usando la regla de la cadena
# P(B) = P(B | A) * P(A) + P(B | no A) * P(no A)
P_no_A = 1 - P_A  # Probabilidad de que no llueva

P_B = P_B_dado_A * P_A + P_B_dado_no_A * P_no_A

# Calcular P(A ∩ B) usando la regla de la cadena
P_A_interseccion_B = P_B_dado_A * P_A

# Imprimir resultados
print(f"P(A): {P_A:.2f}")
print(f"P(B | A): {P_B_dado_A:.2f}")
print(f"P(B | no A): {P_B_dado_no_A:.2f}")
print(f"P(B): {P_B:.2f}")
print(f"P(A ∩ B): {P_A_interseccion_B:.2f}")

# Verificación de la regla del producto
# P(A ∩ B) = P(A) * P(B | A)
P_A_interseccion_B_verificacion = P_A * P_B_dado_A
print(f"P(A ∩ B) usando la regla del producto: {P_A_interseccion_B_verificacion:.2f}")
```

## 6.	Variable aleatoria.

Una **variable aleatoria** es una función que asigna un valor numérico (real) a cada posible resultado de un experimento aleatorio. Esta función está definida sobre un espacio muestral.

El espacio muestral $\Omega$ es el conjunto de todos los posibles resultados de un experimento aleatorio. Una variable aleatoria \(X\) es una función que toma cada elemento de $\Omega$ y lo asocia a un número real:

$$
f: \ \Omega \ \rightarrow \mathbb{R}
$$

Esto significa que \(f\) es una función que asigna un número real a cada posible resultado del experimento. El valor que \(f\) toma depende del resultado aleatorio que se observe en $\Omega$.

Suponiendo que $\Omega$ consta de todos los lanzamientos de una moneda en 10 ocasiones, entonces

$$
\omega \in \{AA...A, SA...A, ..., SS\}
$$

Con esto es más fácil hablar de la medida de probabilidad de un evento que involucra una variable aleatoria.


### Ejemplo 6.1

```python
import numpy as np
import itertools
import matplotlib.pyplot as plt

# Definir el espacio muestral para 10 lanzamientos de una moneda (Águila = A, Sol = S)
n_lanzamientos = 10
moneda = ['A', 'S']

# Obtener todas las posibles combinaciones de resultados para los 10 lanzamientos
espacio_muestral = list(itertools.product(moneda, repeat=n_lanzamientos))

# Definir la variable aleatoria X como el número de Águilas en cada resultado
def variable_aleatoria(omega):
    return omega.count('A')

# Calcular X para todos los elementos del espacio muestral
valores_X = np.array([variable_aleatoria(omega) for omega in espacio_muestral])

# Visualizar la distribución de la variable aleatoria (Número de Águilas)
valores_unicos, cuentas = np.unique(valores_X, return_counts=True)

# Graficar la distribución de probabilidad
plt.bar(valores_unicos, cuentas / len(espacio_muestral))
plt.title('Distribución de la Variable Aleatoria X (Número de Águilas en 10 lanzamientos)')
plt.xlabel('Número de Águilas (X)')
plt.ylabel('Probabilidad')
plt.grid(True)
plt.show()

# Mostrar algunos ejemplos del espacio muestral y sus valores de X
print(f"Ejemplo de espacio muestral (primeros 5 resultados): {espacio_muestral[:5]}")
print(f"Valores de X para los primeros 5 resultados: {valores_X[:5]}")
```

## 7.	Distribución de probabilidad.

Una distribución de probabilidad es una función que describe la probabilidad de que un evento aleatorio tome ciertos valores, define cómo se distribuyen las probabilidades a lo largo de los posibles resultados de un experimento aleatorio.

### Ejemplo 7.1


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# ----------------------------
# Distribución de Probabilidad Discreta: Distribución Binomial
# ----------------------------
n = 10  # Número de ensayos (lanzamientos)
p = 0.5  # Probabilidad de éxito (Águila)
x = np.arange(0, n+1)  # Posibles valores de éxitos (0 a 10)

# Distribución binomial: Probabilidad de obtener 'k' águilas en 'n' lanzamientos
binom_pmf = binom.pmf(x, n, p)

# Graficar la distribución binomial
plt.bar(x, binom_pmf, color='skyblue', label=f'Binomial(n={n}, p={p})')
plt.title('Distribución Binomial - Probabilidad de Éxito en 10 Lanzamientos')
plt.xlabel('Número de Éxitos (Águilas)')
plt.ylabel('Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# ----------------------------
# Distribución de Probabilidad Continua: Distribución Normal
# ----------------------------
mu = 0  # Media
sigma = 1  # Desviación estándar
x_cont = np.linspace(-4, 4, 1000)

# Distribución normal: Densidad de probabilidad
normal_pdf = norm.pdf(x_cont, mu, sigma)

# Graficar la distribución normal
plt.plot(x_cont, normal_pdf, color='purple', label=f'Normal(μ={mu}, σ={sigma})')
plt.title('Distribución Normal - Densidad de Probabilidad')
plt.xlabel('Valor de X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

## 8.	Distribución Binomial.

La **distribución binomial** es un modelo de probabilidad que describe el número de éxitos en una secuencia de ensayos independientes y con dos resultados posibles: éxito o fracaso. Esta distribución de probabilidad mide la probabilidad del número de resultados \(k\) favorables para un evento en \(n\) experimentos que son independientes entre sí.

### Función de masa de probabilidad

La función de masa de probabilidad se expresa como:

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

donde \(P(X = k)\) es la probabilidad de obtener exactamente \(k\) éxitos en \(n\) ensayos.

El término $$\binom{n}{k}$$ es el **coeficiente binomial**, que calcula las combinaciones de \(n\) elementos tomados de \(k\) en \(k\):

$$
\binom{n}{k} = \frac{n!}{k! \ (n-k)!}
$$

Aquí, $$p^k$$ es la probabilidad de tener \(k\) éxitos y $$(1-p)^{n-k}$$ es la probabilidad de tener \(n-k\) fracasos.


### Ejemplo 8.1


```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb
from scipy.stats import binom

# Definir parámetros de la distribución binomial
n = 10  # Número de ensayos (lanzamientos)
p = 0.5  # Probabilidad de éxito (Águila)
k_values = np.arange(0, n+1)  # Posibles valores de éxitos (0 a n)

# Función de masa de probabilidad (PMF) - fórmula manual
def binomial_pmf(k, n, p):
    return comb(n, k) * (p**k) * ((1-p)**(n-k))

# Calcular la PMF usando la fórmula manual
pmf_manual = [binomial_pmf(k, n, p) for k in k_values]

# Calcular la PMF usando la función binom.pmf de scipy
pmf_scipy = binom.pmf(k_values, n, p)

# Graficar la PMF calculada manualmente
plt.bar(k_values, pmf_manual, color='skyblue', label='PMF (Manual)', alpha=0.6)

# Graficar la PMF calculada con scipy
plt.plot(k_values, pmf_scipy, 'ro-', label='PMF (scipy)', alpha=0.8)

plt.title('Distribución Binomial - Número de Éxitos en 10 Lanzamientos')
plt.xlabel('Número de Éxitos (k)')
plt.ylabel('Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# Mostrar los valores de la PMF calculada manualmente y con scipy
for k, pmf_m, pmf_s in zip(k_values, pmf_manual, pmf_scipy):
    print(f"P(X = {k}) = {pmf_m:.4f} (Manual), {pmf_s:.4f} (scipy)")
```

## 9.	Función de distribución.

La **función de distribución** es una función que describe la probabilidad de que una variable aleatoria tome un valor menor o igual a un valor específico. Si \(x\) está definida en $\Omega$ de modo que $x(\omega) = \omega$, entonces:

$$
P\{\omega : x(\omega) \in B\} = P(B) = \int_{B} f(x) \, dx
$$

Esto se puede relacionar con la función de distribución acumulada \(F_X(x)\) porque, en esencia, \(F_X(x)\) nos da la probabilidad acumulada hasta un cierto valor \(x\). Cuando quieres calcular \(P(B)\) para un intervalo, puedes ver a \(B\) como un conjunto que se define entre dos valores \(a\) y \(b\):

$$
P(a < X \leq b) = F_X(b) - F_X(a) = \int_{a}^{b} f(x) \, dx
$$

Además, \(X\) es absolutamente continua si existe una función no negativa \(f = f_X\) definida en \(\mathbb{R}\) tal que

$$
F_X(x) = \int_{-\infty}^{x} f_X(t) \, dt \quad \forall \, x \in \mathbb{R}
$$


### Ejemplo 9.1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definir los parámetros de la distribución normal
mu = 0  # Media
sigma = 1  # Desviación estándar

# Crear un rango de valores para x
x = np.linspace(-5, 5, 1000)

# Función de densidad de probabilidad (PDF) para la distribución normal
pdf = norm.pdf(x, mu, sigma)

# Función de distribución acumulada (CDF) para la distribución normal
cdf = norm.cdf(x, mu, sigma)

# Graficar PDF y CDF
plt.figure(figsize=(10, 6))

# Graficar PDF
plt.subplot(2, 1, 1)
plt.plot(x, pdf, 'b', label='PDF')
plt.title('Función de Densidad de Probabilidad (PDF)')
plt.ylabel('Densidad')
plt.grid(True)

# Graficar CDF
plt.subplot(2, 1, 2)
plt.plot(x, cdf, 'r', label='CDF')
plt.title('Función de Distribución Acumulada (CDF)')
plt.ylabel('Probabilidad acumulada')
plt.grid(True)

plt.tight_layout()
plt.show()

# Cálculo de P(a < X <= b)
a, b = -1, 1
probabilidad = norm.cdf(b, mu, sigma) - norm.cdf(a, mu, sigma)
print(f"P({a} < X <= {b}) = {probabilidad:.4f}")
```

## 10.	Funciones de densidad conjunta.

En el caso que necesitamos asignar probabilidades basadas en $\omega$ área, entonces el espacio de probabilidad correspondiente involucra dos variables aleatorias de manera simultánea:

$$
P\{(x,y) \in B\} = P(B) = \int_{B} f_{XY}(x,y) \, dx \, dy
$$

Si $X$ y $Y$ son dos variables aleatorias continuas, la función de densidad conjunta $f_{XY}(x,y)$ es una función que describe la probabilidad de que $X$ y $Y$ caigan en un cierto conjunto $B$ en el plano $XY$.

La función $f_{XY}(x,y)$ no se interpreta como una probabilidad en sí, sino como una densidad. Para que esta función sea válida, debe cumplir con las siguientes propiedades:

$ f_{XY}(x,y) \geq 0 $ para todos $x$ y $y$

$$
\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f_{XY}(x,y) \, dx \, dy = 1
$$


### Ejemplo 10.1


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Parámetros de la distribución normal bivariada
mu = [0, 0]  # Medias de X y Y
cov = [[1, 0.5], [0.5, 1]]  # Matriz de covarianza (relación entre X y Y)

# Crear una cuadrícula de valores para X y Y
x = np.linspace(-3, 3, 500)
y = np.linspace(-3, 3, 500)
X, Y = np.meshgrid(x, y)

# Evaluar la función de densidad conjunta (PDF conjunta)
pos = np.dstack((X, Y))  # Crear una matriz de coordenadas (x, y)
rv = multivariate_normal(mu, cov)  # Distribución normal bivariada
Z = rv.pdf(pos)  # Función de densidad conjunta

# Graficar la función de densidad conjunta
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie 3D
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Etiquetas y título
ax.set_title('Función de Densidad Conjunta (Distribución Normal Bivariada)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Densidad')

plt.show()

# Cálculo de P{(X,Y) ∈ B} para un conjunto B definido por -1 ≤ X ≤ 1 y -1 ≤ Y ≤ 1
probabilidad = rv.cdf([1, 1]) - rv.cdf([-1, -1])
print(f"P{{-1 ≤ X ≤ 1, -1 ≤ Y ≤ 1}} = {probabilidad:.4f}")
```

## 11.	Función de probabilidad de un evento conjunto (bidimensional).

Es un concepto que se refiere a la probabilidad de que dos variables aleatorias tomen simultáneamente ciertos valores dentro de un conjunto bidimensional. Esta probabilidad se calcula utilizando la función de densidad conjunta $f_{XY}(x,y)$, la cual describe cómo se distribuyen las probabilidades sobre un espacio bidimensional formado por dos variables aleatorias $X$ y $Y$.

$$
P\{(x,y) \in B\} = \int_{B} f_{XY}(x,y) \, dx \, dy
$$

La función de probabilidad conjunta en un espacio n-dimensional es una extensión natural del caso bidimensional, donde involucramos $n$ variables aleatorias simultáneamente. La probabilidad total de que las $n$ variables aleatorias tomen cualquier valor en el espacio n-dimensional es igual a 1:

$$
\int_{-\infty}^{\infty} \cdots \int_{-\infty}^{\infty} f_{X_1,X_2,\ldots,X_n}(x_1,x_2,\ldots,x_n) \, dx_1 \, dx_2 \cdots dx_n = 1
$$

Lo que prueba que $f_{X_1,X_2,\ldots,X_n}$ es una función de densidad y, por lo tanto, $F_{X_1,X_2,\ldots,X_n}$ es una distribución de probabilidad conjunta de n-dimensiones.


### Ejemplo 11.1

```python
import itertools
import numpy as np
import pandas as pd

# Definimos el espacio muestral Omega para lanzar una moneda 10 veces
moneda = ['A', 'S']  # A = Águila, S = Sol
n_lanzamientos = 10

# Generamos todas las combinaciones posibles de lanzamientos
espacio_muestral = list(itertools.product(moneda, repeat=n_lanzamientos))

# Definimos una variable aleatoria que cuenta el número de caras (águilas)
def variable_aleatoria(resultado):
    return resultado.count('A')

# Asignamos valores numéricos a cada resultado
valores = [variable_aleatoria(resultado) for resultado in espacio_muestral]

# Creamos un DataFrame para mostrar resultados y sus valores
df = pd.DataFrame({
    'Resultado': [''.join(resultado) for resultado in espacio_muestral],
    'Número de Águilas': valores
})

# Mostramos los primeros 10 resultados
print(df.head(10))

# Resumen de la variable aleatoria
resumen = df['Número de Águilas'].value_counts().sort_index()
print("\nResumen de la variable aleatoria:")
print(resumen)
```

## 12.	Calculo de la densidad marginal en n-dimensiones.

Para el caso de **variables aleatorias continuas**, tanto la regla de la suma como el teorema de la probabilidad total tienen análogos, pero en lugar de trabajar con sumas (como en el caso discreto), trabajamos con integrales.

### Regla de la suma

La regla de la suma en el caso continuo se aplica cuando estamos interesados en calcular la probabilidad de un evento relacionado con una variable aleatoria continua, integrando la función de densidad marginal o conjunta sobre diferentes subconjuntos.

$$
f_X(x) = P\{x \in B\} = \int_{-\infty}^{\infty} f_{XY}(x,y) \, dy
$$

$$
f_Y(y) = P\{y \in B\} = \int_{-\infty}^{\infty} f_{XY}(x,y) \, dx
$$

Para $n$ dimensiones:

$$
\int_{-\infty}^{\infty} \cdots \int_{-\infty}^{\infty} f(x_1, \ldots, x_n) \, dx_1 \cdots dx_n = f(x_i)
$$



### Ejemplo 12.1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import dblquad

# Definimos la función de densidad conjunta f_{XY}(x, y)
def f_xy(x, y):
    return (1 / 8) * (2 + x + y)  # Ejemplo de función de densidad conjunta

# Calculamos la densidad marginal f_X(x) integrando sobre y
def f_x(x):
    # Integrar f_{XY}(x, y) respecto a y desde -1 a 1 (o el rango apropiado)
    return dblquad(f_xy, -1, 1, lambda y: -1, lambda y: 1)[0]

# Valores de x para calcular f_X(x)
x_values = np.linspace(-1, 1, 100)
f_x_values = np.array([f_x(x) for x in x_values])

# Graficamos la densidad marginal f_X(x)
plt.figure(figsize=(10, 6))
plt.plot(x_values, f_x_values, label='Densidad Marginal f_X(x)', color='blue')
plt.title('Densidad Marginal de X')
plt.xlabel('x')
plt.ylabel('f_X(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```

## 13.	Independencia de variables aleatorias.

La **independencia** entre variables aleatorias significa que el conocimiento sobre el valor de una variable no proporciona ninguna información sobre los valores de las demás variables. Formalmente, se dice que un conjunto de variables aleatorias $X_1, X_2, \ldots, X_n$ son independientes si la probabilidad conjunta de que cada una de las variables caiga en un conjunto específico es igual al producto de las probabilidades individuales de que cada variable caiga en su respectivo conjunto.

Hasta ahora se ha considerado la independencia entre eventos. Se puede tomar la analogía para definir la independencia entre variables aleatorias.

Esto es: Sean $X_1, \ldots, X_n$ variables aleatorias, estas son independientes entre sí; el conocimiento que se tiene sobre cualquier \(X_i\) no cambia el que se tiene sobre las otras $X_j$.

Formalmente, $X_1, \ldots, X_n$ son independientes si para todos los conjuntos $B_1, \ldots, B_n$ se cumple que

$$
P\{X_1 \in B_1, \ldots, X_n \in B_n\} = P\{X_1 \in B_1\} \cdot \ldots \cdot P\{X_n \in B_n\}
$$


### Ejemplo 13.1


import numpy as np

mean_X, std_X = 0, 1  # Media y desviación estándar de X

# Generar muestras independientes de X y Y

## 14.	Valor esperado.

El **valor esperado** (o esperanza matemática) de una variable aleatoria representa el "promedio ponderado" de todos los posibles valores que puede tomar la variable, ponderado según la probabilidad de que esos valores ocurran.

Si \(X\) es una variable aleatoria discreta que puede tomar los valores \(x_1, x_2, \ldots, x_n\) y \(P(X = x_i)\) es la probabilidad de que \(X = x_i\), el valor esperado de \(X\) se define como:

$$
E[X] = \sum_{i} x_i \cdot P(X = x_i)
$$

Si \(X\) es una variable aleatoria continua con función de densidad de probabilidad \(f_X(x)\), el valor esperado de \(X\) se define como:

$$
E[X] = \int_{-\infty}^{\infty} x \cdot f_X(x) \, dx
$$

### Propiedades de \(E[X]\)

1. El valor esperado es lineal, lo que significa que para dos variables aleatorias \(X\) y \(Y\), y constantes \(a\) y \(b\), se cumple lo siguiente:

$$
E[aX + bY] = aE[X] + bE[Y]
$$

2. Si \(g(X)\) es una función de una variable aleatoria \(X\), el valor esperado de \(g(X)\) se calcula usando la distribución de \(X\):

$$
E[g(X)] = \sum_{i} g(x_i) \cdot P(X = x_i)
$$

3. Si \(X\) y \(Y\) son variables aleatorias independientes, el valor esperado del producto de \(X\) y \(Y\) es igual al producto de sus valores esperados:

$$
E[XY] = E[X] \cdot E[Y]
$$


### Ejemplo 14.1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Ejemplo 1: Caso Discreto
# Valores que X puede tomar y sus respectivas probabilidades
valores_discretos = np.array([1, 2, 3])
probabilidades = np.array([0.2, 0.5, 0.3])

# Calcular el valor esperado en el caso discreto
valor_esperado_discreto = np.sum(valores_discretos * probabilidades)

print(f"Valor esperado para el caso discreto: E[X] = {valor_esperado_discreto}")

# Ejemplo 2: Caso Continuo
# Definimos la función de densidad de probabilidad (Distribución Normal)
media, desviacion_estandar = 0, 1  # Normal estándar

# Definimos la función de densidad
f_X = lambda x: norm.pdf(x, loc=media, scale=desviacion_estandar)

# Generamos valores para x
x_values = np.linspace(-4, 4, 1000)

# Calcular el valor esperado en el caso continuo
valor_esperado_continuo = np.trapz(x_values * f_X(x_values), x_values)

print(f"Valor esperado para el caso continuo: E[X] = {valor_esperado_continuo}")

# Graficamos la distribución continua
plt.figure(figsize=(8, 5))
plt.plot(x_values, f_X(x_values), label=f'Densidad de X (Normal)')
plt.axvline(x=valor_esperado_continuo, color='red', linestyle='--', label=f'E[X] = {valor_esperado_continuo:.2f}')
plt.title('Función de Densidad de X y su Valor Esperado')
plt.xlabel('x')
plt.ylabel('f_X(x)')
plt.legend()
plt.grid()
plt.show()
```

## 15.	Momento de una variable aleatoria.

El **momento** de una variable aleatoria es una medida estadística que describe características importantes de la distribución de la variable. En general, los momentos se utilizan para capturar diversas propiedades de una distribución, como su centro, dispersión y forma.

El \(n\)-ésimo momento de una variable aleatoria \(X\) es el valor esperado de $X^n$. El momento de orden \(n\) se denota por $E[X^n]$.

### Para variables discretas:

$$
E[X^n] = \sum_{i} x_i^n \cdot P(X = x_i)
$$

### Para variables continuas:

$$
E[X^n] = \int_{-\infty}^{\infty} x^n \cdot f_X(x) \, dx
$$


### Ejemplo 15.1


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.integrate import quad

# Función para calcular el n-ésimo momento en el caso discreto
def momento_discreto(valores_discretos, probabilidades, n):
    return np.sum(valores_discretos**n * probabilidades)

# Función para calcular el n-ésimo momento en el caso continuo
def momento_continuo(f_X, n):
    # Integración numérica del momento
    return quad(lambda x: x**n * f_X(x), -np.inf, np.inf)[0]

# Ejemplo 1: Caso Discreto
valores_discretos = np.array([1, 2, 3, 4])
probabilidades = np.array([0.1, 0.3, 0.4, 0.2])
n_discreto = 2  # Calculamos el segundo momento (varianza)
momento_2_discreto = momento_discreto(valores_discretos, probabilidades, n_discreto)

print(f"Segundo momento para el caso discreto (n=2): E[X^2] = {momento_2_discreto}")

# Ejemplo 2: Caso Continuo
# Definimos la función de densidad (Distribución Normal)
media, desviacion_estandar = 0, 1  # Normal estándar
f_X = lambda x: norm.pdf(x, loc=media, scale=desviacion_estandar)

# Calculamos el segundo momento para el caso continuo
n_continuo = 2  # Segundo momento (varianza)
momento_2_continuo = momento_continuo(f_X, n_continuo)

print(f"Segundo momento para el caso continuo (n=2): E[X^2] = {momento_2_continuo}")

# Graficamos la función de densidad continua
x_values = np.linspace(-4, 4, 1000)
plt.plot(x_values, f_X(x_values), label='f_X(x) (Normal)')
plt.title('Función de Densidad de X (Normal)')
plt.xlabel('x')
plt.ylabel('f_X(x)')
plt.legend()
plt.grid()
plt.show()
```
